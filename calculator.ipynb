```python
5# Import 'math' and 'random' Libraries
import math
import random
# Define Variables
output = 0
num1 = ""
operator = ""
num2 = ""


# Define Function Listing Function
def abilitiesList():
    print("+...Addition")
    print("-...Subtractïœ‰ion")
    print("*...Multiplication")
    print("/...Division")
    print("^...Powers")
    print("/-...Square Roots ")
    print("!...Factorials (Input Cannot Be Negetave)")
    print("fib...Fibonacci Series")
    print("mean...Mean ")
    print("median...Median")
    print("pnc...Permutation")
    print("mode...Mode")
    print("var...Variance")
    print("stdv...Standard Deviation")
    print("com...Combination")
    print("rev...Reverse a number")
    print("and...And Operator")
    print("or...Or Operator")
    print("//////////////////////////////////////////////////////////////////////////")

def askForNumInput(textPrompt):
    # Devine local variable
    convertedNum = math.nan

    # Wait for valid numerical input 
    while True:
        num = input(textPrompt)
        try:
            # Try to typecast the input to a float
            float(num)
        except ValueError:
            # Catch the exception if it is not a number
            print("ERROR: Syn: Invalid Num")
        else:
            # Typecasting
            convertedNum = float(num)
            break
    return convertedNum

# While Loop
while True:
  
    print("//////////////////////////////////////////////////////////////////////////")
    print("Type 'help' for a list of abilities")
    # Loop for getting operation
    while True:
      
        operator = input("What operation do you want to perform? ")
        # Is operator == to any of out constants or predefines?
        if operator == "help":
            abilitiesList()
        
        # Has the user entered in a valid operator?
        elif operator == "+" or operator == "-" or operator == "*" or operator == "/" or operator == "^" or operator == "/-" or operator == "!" or operator == "fib" or operator == "and" or operator == "or" or operator=="rev"  or operator == "mean" or operator == "median" or operator == "pnc" or operator == "mode" or operator == "var" or operator == "stdv" or operator == "com" or operator == "log":
            break
        else:
            print("ERROR: Invalid Operator")

    # Loop for 1st number input
    while True:
        num1 = askForNumInput("First Number? ")
        
        
        if operator == "!" and num1 < 0:
            print("ERROR: Math: Factorials only accept inputs > 0")
        elif operator == "Fib" and num1 < 1:
            print("ERROR: Math: Factorials only accept inputs > 0")

        else:
          break

    # Does the operation require a 2nd input?
    if not (operator=="/-" or operator=="!" or operator=="fib" or operator=="rev" or operator=="mean" or operator=="median" or operator=="pnc" or operator=="mode" or operator=="var" or operator=="stdv" or operator=="com"):
        # Loop for 2nd number input
        while True:
            num2 = askForNumInput("Second Number? ")
            # Catch x/0 error case
            if  operator == "/" and num2 == "0":
                print("ERROR: Math: Canot divide by 0!")
            else:
                break

    # Calculations
    if operator == "+":
        output = num1 + num2
        print("Your Answer: "+str(output))
    elif operator == "-":
        output = num1 - num2
        print("Your Answer: "+str(output))
    elif operator == "*":
        output = num1 * num2
        print("Your Answer: "+str(output))
    elif operator == "/":
        output = num1 / num2
        print("Your Answer: "+str(output))
    elif operator == "^":
        output = math.pow(num1,num2)
        print("Your Answer: "+str(output))
    elif operator == "/-":
        output = math.sqrt(num1)
        print("Your Answer: "+str(output))
    elif operator == "!":
        output = math.factorial(num1)
        print("Your Answer: "+str(output))
    elif operator == "fib":
      n1, n2 = 0, 1
      count = 0
      if num1 <= 0:
        print("Please enter a positive integer")
      elif num1 == 1:
        print("Fibonacci sequence upto",num1,":")
        print(n1)
      else:
        print("Fibonacci sequence:")
        while count < num1:
          print(n1)
          nth = n1 + n2
          n1 = n2
          n2 = nth
          count += 1        
        print("Completed")
    elif operator == "mean":
      import statistics
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      x = statistics.mean(a)
      print("Mean is :", x)
    elif operator == "median":
      import statistics
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      x = statistics.median(a)
      print("Median is :", x)
    elif operator == "mode":
      import statistics
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      x = statistics.mode(a)
      print("mode is :", x)
    elif operator == "var":
      import statistics
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      x = statistics.variance(a)
      print("variance is :", x)
    elif operator == "stdv":
      import statistics
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      x = statistics.stdev(a)
      print("stdev is :", x)
    elif operator == "com":
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      n = int(input("Enter length of combination"))
      print( a)
      from itertools import combinations
      num = combinations(a, n)  
      print(num)  
      for p in list(num):
        print(p) 
    elif operator == "pnc":
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      print( a)
      from itertools import permutations
      num = permutations(a)  
      print(num)  
      for p in list(num):
        print(p) 
    elif operator == "rev":
      a = list(map(int,input("\nEnter the numbers : ").strip().split()))[:]
      print(a)
      print(str(a[::-1]))
    elif operator == "and":
      num = eval(input("Enter number greaer than first And Less than second no: "))
      if num > num1 and num < num2:
        print("True:",num, "is greater than", num1, "And", num, "is less than", num2)
      else:
        print("False: Both arguments are not true")
    elif operator == "or":
      num = eval(input("Enter number greaer than first or Less than second no: "))
      if num > num1 or num < num2:
        print("True:",num, "is greater than", num1, "Or", num, "is less than", num2)
      else:
        print("False: Atleast one arguments is not true")
```
